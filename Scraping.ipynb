```python
#IMPORTING LIBRARIES
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import re
```

```python
#SETTING UP WEBDRIVER
chrome_options = Options()
chrome_options.add_argument("--no-sandbox") 
chrome_options.add_argument("--auto-open-devtools-for-tabs")
driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
```

```python
#NAVIGATING TO THE WEBSITE
driver.get("https://games.crossfit.com/leaderboard/open/2020?view=0&division=1&scaled=0&sort=0")
```

```python
#RECEIVING NAMES
names = driver.find_elements_by_class_name("name")
names = [name.text for name in names]
hidden_info = driver.find_elements_by_class_name("athlete.collapsed")
for i in hidden_info:
    i.click()
```

```python
#INFO FOR EVENTS
lst = []
for i in hidden_info:
    for j in i.find_elements_by_class_name("result"):
        lst.append(j.text)
stats = [lst[x:x+5] for x in range(0, len(lst), 5)]
```

```python
#GETTING DATA INTO NEATER FORMAT
count = 0
for i in stats:
    count2 = 0
    for j in i:
        j = j.replace("(", "").replace(")", "")
        stats[count][count2] = j
        count2 += 1
    count += 1
```

```python
#SCROLLING BACK TO TOP OF SCREEN
import time
driver.execute_script("window.scrollTo(300, document.body.scrollHeight);")
info = driver.find_elements_by_class_name("profile")
for i in info:
    i.click()
```

```python
# MORE CLEANING
info = driver.find_elements_by_class_name("info")
info = [i.text for i in info]
edit = [i.split("\n") for i in info]
```

```python
# MORE CLEANING
for i in range(len(edit)):
    edit[i][2] = edit[i][2].split(" | ")
```

```python
#ASSEMBLING ALL INFO INTO LISTS
all_stats = []
for i in range(len(edit)):
    all_stats.append([])
for i in range(len(all_stats)):
    for j in edit[i]:
        print(j)
        if isinstance(j, list):
            for k in j:
                all_stats[i].append(k)
            continue;   
        all_stats[i].append(j)
        
```

```python
#GETTING RANK OF EACH ATHLETE
rank = driver.find_elements_by_class_name("pos")
rank = [i.text for i in rank]
```

```python
# FINAL STEP TO HAVE EACH ATHLETES INFORMATION IN A LIST
final = []
names = [[name] for name in names]
for i in range(len(all_stats)):
    names[i].extend(all_stats[i])
    names[i].extend(stats[i])
    names[i].append(rank[i])
    
```

```python
# DOING THIS 100 TIMES, OR HOWEVER MANY TIMES DEPENDING ON HOW MUCH DATA YOU WANT
for i in range(100):
    
    # SCROLLNG TO RIGHT LOCATION
    driver.execute_script("window.scrollTo(0, 12000)")
    next_ = driver.find_element_by_class_name("btn.next")
    #CLICKING ON INFORMATION SO IT WILL BE AVAILABLE TO SCRAPE
    next_.click()
    time.sleep(4)
    
    #GETTING NAMES, AND ASSOCIATED INFO 
    names = driver.find_elements_by_class_name("name")
    names = [name.text for name in names]
    hidden_info = driver.find_elements_by_class_name("athlete.collapsed")
    for i in hidden_info:
        try:
            i.click()
        except:
            pass
    lst = []
    
    # GRABBING RESULTS FOR EACH EVENT
    for i in hidden_info:
        for j in i.find_elements_by_class_name("result"):
            lst.append(j.text)
    stats = [lst[x:x+5] for x in range(0, len(lst), 5)]
    count = 0
    for i in stats:
        count2 = 0
        for j in i:
            j = j.replace("(", "").replace(")", "")
            stats[count][count2] = j
            count2 += 1
        count += 1
    
    #GRABBING THE RANKING
    rank = driver.find_elements_by_class_name("pos")
    rank = [i.text for i in rank]
    
    info = driver.find_elements_by_class_name("profile")

    #IF ELEMENT IS UNCLICKABLE, JUST KEEP GOING, DON'T POP UP AN ERROR
    for i in info:
        try:
            i.click()
        except:
            pass
    
    #SCROLL BACK TO TOP, COLLECT MORE INFO
    driver.execute_script("window.scrollTo(300, document.body.scrollHeight);")
    info = driver.find_elements_by_class_name("info")
    info = [i.text for i in info]
    all_stats = [i.split("\n") for i in info]
    
    #MORE CLEANING OF THE DATA THAT WAS SCRAPED
    for i in range(len(all_stats)):
        try:
            all_stats[i][2] = all_stats[i][2].split(" | ")
        except:
            pass
    
    # MAKING SURE ALL THE DATA IS PLACED NICELY IN A ONE DIMENSIONAL ARRAY
    update = []
    for i in range(len(all_stats)):
        update.append([])
    for i in range(len(update)):
        for j in all_stats[i]:
            if isinstance(j, list):
                for k in j:
                    update[i].append(k)
                continue;   
            update[i].append(j)
    
    #MAKING THE FINAL LIST OF LISTS FOR 50 ATHLETES ON EACH PAGE
    final = []
    names = [[name] for name in names]

    for i in range(len(update)):
        names[i].extend(update[i])
        names[i].extend(stats[i])
        names[i].append(rank[i])

    #EACH LOOP WILL ADD 50 MORE ROWS TO THE EXISTING DATAFRAME CREATED BEORE THIS LOOP
    df2 = pd.DataFrame(names)  
    df = pd.concat([df, df2], axis = 0, ignore_index = True)

```

```python
# SAVING DF EVERY 5000 ROWS, THIS PARTICULAR FILE HAS 20,000 ROWS
df.to_csv("fourth.csv", index = False)
```
